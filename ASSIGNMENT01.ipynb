# 5 

Perform meaningful visualizations using the dataset.

# a.

```python
# using seaborn displaying the graph
sns.countplot(data,x='Classification')

#using matplotlib,

#set title of the graph,size and colour
plt.title('DATA OF CLASSIFICATIONS',size=20,color='r')

#set xlabel and ylabel of the graph
plt.xlabel('CLASSIFICATIONS')
plt.ylabel('COUNT')

#show the plot
plt.show()
```

# b.

```python
# using matplotlib,

#displaying the graph with classification and sl
plt.bar(data['Classification'],data['SL'])


#set title of the graph with size and colour
plt.title('BAR GRAPH OF SEPAL LENGTH',size=20,color='r')

#set xlabel and ylabel of the graph
plt.xlabel('CLASSIFICATION OF SPECIES')
plt.ylabel('SEPAL LENGTH (SL)')

#show the plot
plt.show()
```

# C.

```python
#using matplotlib,

#plot the figure with a size 
plt.figure(figsize=(10,4))

#displaying the sepal width as histogram with colour and width 
plt.hist(data['SW'],color='y',rwidth=.8)

#set title of the graph with size and colour
plt.title('HISTOGRAM OF SEPAL WIDTH',size=25,color='g')

#set xlabel and ylabel of the graph
plt.xlabel('SW')
plt.ylabel('count')

#show the plot
plt.show()
```

# d.

```python
#create boxplot for petal length
plt.boxplot(data['PL'])

#set title of the plot with size and colour
plt.title('BOXPLOT OF PETAL LENGTH',size=18,c='y')

#show the plot
plt.show()
```

# e.

```python
#by using seaborn pairplot the data
sns.pairplot(data)

#show the plot
plt.show()
```

# f.

```python
#by using seaborn create a boxplot of sl,sw,pl,pw
sns.boxplot(data)

#show the plot
plt.show()
```

# g.

```python
# by using seaborn creating a boxplot of petal width 
sns.boxplot(x='Classification',y='PW',data=data)

#show the plot
plt.show()
```

# h.

```python
#Creating a scatter plot of sepal length with size,marker & colour
plt.scatter(data['Classification'],data['SL'],s=10,marker='x',color='b')

#set title of the scatter plot with size and colour
plt.title('SCATTER PLOT OF SEPAL LENGTH',size=18,c='g')

#show the plot
plt.show()
```
